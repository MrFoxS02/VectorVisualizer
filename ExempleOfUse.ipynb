```python
dino_vector = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

dino_graph = draw_2D_poligon()

def add(v1,v2):
    return (v1[0] + v2[0], v1[1] + v2[1])

dino_vector2 = [add((-1.5, -2.5), v) for v in dino_vector]
dino_graph.get_vector(dino_vector)
dino_graph.draw_poligon(color='b')
dino_graph.get_vector(dino_vector2)
dino_graph.draw_poligon(color='r')
```

```python
vect = vector_2D_transform()
dinos_y = [[vect.add((12 * i, 10 * j), v) for v in dino_vector] for i in range(-10, 10) for j in range(-10, 10)]
```

```python
plt.figure(figsize=(20, 20))
dino_graph = draw_2D_poligon()
for p in dinos_y:
    dino_graph.get_vector(p, ax_padding=0, ax_lim=[5, 5, 5, 5])
    dino_graph.draw_poligon()
```

```python
polar = [(math.cos(5*x*math.pi/500), 2* math.pi *x/1000) for x in range(1000)]
```

```python
vectors = [vect.to_cartesian(p) for p in polar]
```

```python
dino_graph = draw_2D_poligon()
dino_graph.get_vector(vectors, ax_padding=0, ax_lim=[0.2, 0.2, 0.2, 0.3])
dino_graph.draw_poligon(color='green', if_axis=True, if_grid=True)
```

```python
new_dino = vect.translate((8, 8), vect.rotate(dino_vector, 5 * math.pi / 3))
```

```python
dino_graph = draw_2D_poligon()
dino_graph.get_vector(dino_vector)
dino_graph.draw_poligon(color='red', if_axis=False, if_grid=False)
rot_d = vect.rotate(dino_vector, 5*math.pi/3)
trans_d = vect.translate((8, 8), rot_d)
dino_graph.get_vector(trans_d)
dino_graph.draw_poligon(color='green', if_axis=False, if_grid=False)
dino_graph.get_vector(new_dino)
dino_graph.draw_poligon(color='blue', if_axis=False, if_grid=False)
dino_graph.get_vector(rot_d)
dino_graph.draw_poligon(color='orange', if_axis=True, if_grid=True)
```
